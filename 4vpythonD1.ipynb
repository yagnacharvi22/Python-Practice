```python
#ATM problem
amount=int(input("Enter the amount"))
fiv5=0
tw2=0
on1=0
if amount%100==0:
  if amount>=500:
    fiv5=amount//500
    amount=amount-(fiv5*500)
    print("500 ------- ",fiv5)
  if amount>=200:
    tw2=amount//200
    amount=amount-(tw2*200)
    print("200 ------- ",tw2)
  if amount>=100:
    on1=amount//100
    amount=amount-(on1*100)
    print("100 ------- ",on1)
else:
    print("invalid amount")
```

```python
#accept project marks, internal marks , external marks from student
#pass marks is 50 in each subject
#calculate total score if a student got more than or equal to 50 in each subject.
#total_score= 70% on project + 20% on external + 10% on internal
#print A grade if score is 90+
#print B grade if score is b/w 70-90
#C grade if score is b/w 50-70

project=int(input("Enter the Project Marks"))
internal=int(input("Enter the Internal Marks"))
external=int(input("Enter the External Marks"))
total_score=(project*0.7)+(internal*0.1)+(external*0.2)
print("Total Score",total_score)
if total_score>=90:
  print("Grade A")
elif total_score>=70:
  print("Grade B")
elif total_score>=50:
  print("Grade C")
else:
  print("Fail")


```

```python
start=int(input("Enter the start value"))
stop=int(input("Enter the end value"))
if start<stop:
  for i in range(start,stop):
    print(i, end=" ")
elif start>stop:
  for i in range(stop,start,-1):
    print(i, end=" ")
else:
  print("Your start and stop values are same")
```
